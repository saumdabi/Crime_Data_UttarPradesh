```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv("../input/up-crime-data-year-2019/MR data - Compiled Data Set.csv")
df=df.fillna(0)
df=df.drop('Title',axis='columns')
df=df.drop('Text',axis='columns')
df=df.replace(["Property Disputes", "Family Dispute", "Petty Quarrels", "Money Disputes",  "Personal Vendetta",  "Love Affairs",  "Casteism", "Unknown reasons" ],[1,2,3,4,5,6,7,8])
g=df.groupby("City ")
df1=g.get_group('Ghaziabad')
df2=g.get_group('Kanpur')
df3=g.get_group('Lucknow')
l1=["child victims","male victims(adult)","female victims(adult)"]
l2=[df1["Number of child victims"].sum(),df1["Number of male victims(adult)"].sum(),df1["Number of female victims(adult)"].sum()]
l3=np.arange(len(l1))
l21=[df2["Number of child victims"].sum(),df2["Number of male victims(adult)"].sum(),df2["Number of female victims(adult)"].sum()]
l22=[df3["Number of child victims"].sum(),df3["Number of male victims(adult)"].sum(),df3["Number of female victims(adult)"].sum()]
df.rename(columns = {'''Crime Against Women ( Combined ):
1. Murder with Rape 
2. Dowry Deaths(Sec. 3048)
3. Suicide(sec 305/306)
4. Kidnapping(All)
5. Acid Attack(Sec. 326A IPC)
6. Cruelty by Husband/in-laws((Sec.498 A IPC)
7. Rape only(Sec. 376 or 511 IPC)
8. Assault on Women with Intent to Outrage her Modesty (Sec. 354 IPC)
9. Cyber Crimes against Women
10. Protection of Children from Sexual Offences Act''':"crimeaw"}, inplace = True)


```
